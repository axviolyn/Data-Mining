```python
# Membaca data

import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DatasetDM/GroceryStoreDataSet.csv', names=['products'], sep=',')
df
```

```python
# Menampilkan data asli

df.values
```

```python
# Transformasi data ke bentuk array

data = list(df["products"].apply(lambda x:x.split(',')))
data
```

```python
# Install library

! pip install mlxtend
```

```python
# Proses dataset dan disimpan dalam dataframe

from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encoded = te.fit(data).transform(data)

df = pd.DataFrame(data_encoded, columns=te.columns_)
df = df.astype(int)
df
```

```python
# Menjalankan fungsi rules

from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```
