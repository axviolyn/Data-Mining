```python
path = '/content/drive/MyDrive/DatasetDM/winequality-red.csv'

# Membaca file csv
import pandas as pd
df = pd.read_csv(path)
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

x = df[["alcohol", "sulphates", "volatile acidity"]].to_numpy()  # Variabel independen
y = df["quality"].to_numpy()  # Variabel dependen

x = sm.add_constant(x)
model_prediksi = sm.OLS(y, x).fit()
print(model_prediksi.summary())
```
