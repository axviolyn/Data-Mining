```python
# Membaca data

import pandas as pd

grocery = pd.read_csv(path + '/GroceryTransactionData.csv', names=['products'], sep=',')
grocery.head()
```

```python
# Menampilkan data asli

grocery.values
```

```python
# Transformasi data ke bentuk array

grocery_transaction = list(grocery["products"].apply(lambda x:x.split(',')))
grocery_transaction
```

```python
# Install library

! pip install mlxtend
```

```python
# Proses dataset dan disimpan dalam dataframe

from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encoded = te.fit(grocery_transaction).transform(grocery_transaction)

df = pd.DataFrame(data_encoded, columns=te.columns_)
df = df.astype(int)
df
```

```python
# Menjalankan fungsi rules

from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
```
